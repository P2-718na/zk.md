```
a = 1/3;
b = 1/2;
V = a*r^3 + b*r^2;

H = p^2 / (2*m) + V
```

```
Plot3D[
    {H /. r->x /. p->y /. m->1},
    {x, -2, 1.5},
    {y, -1, 2},
    BoxRatios->{1.5, 1, 1},
    AxesLabel->{"x", "y", "z"},
    MeshFunctions -> {#3 &}, 
    Mesh -> 40
]
```

# Studio come va l'andamento del potenziale attorno ai punti fissi:

```
Solve[D[V, r] == 0]
```

```
Simplify[(D[V, {r, 2}] /. r->-1) > 0]
```

```
Simplify[(D[V, {r, 2}] /. r->0) > 0]
```

## Le curve di livello sono quindi rappresentate in questo modo:

```
ContourPlot[
    {H /. r->x /. p->y /. m->1},
    {x, -2, 2},
    {y, -2, 2},
    Contours->{
        -1/2,
        0,
        H /. r->-1 /. p->0,
        1/2,
        1,
        3/2,
        2,
        3,
        4
    }
]
```
