# Energia dell'oscillatore iperbolico

```
w = 1;
H[p_, r_] := w/2 * (p^2 - r^2);
```

```
energyPlot = Plot3D[H[p, r], {p, -10, 10}, {r, -10, 10}, MeshFunctions -> {#3 &}, Mesh -> 11]
```

```
origin = Plot3D[0, {p, -10, 10}, {r, -10, 10}, MeshFunctions -> {H[#1, #2] &}, MeshStyle -> Directive[Thick, Red]];
```

```
Show[energyPlot, origin]
```

```
W = 1;
p0 = 1;
alpha = (W^2 + w^2)/W*p0 ;
```

```
x[t_] := p0 * Sinh[w*t]+ alpha/(w^2 + W^2) * (-w*Sin[W*t] + W*Sinh[w*t]);
p[t_] := p0 * Cosh[w*t]+ alpha/(w^2 + W^2) * (-W*Cos[W*t] + W*Cosh[w*t])
```

```
ParametricPlot[{x[t], p[t]}, {t, -5, 5}]
```
