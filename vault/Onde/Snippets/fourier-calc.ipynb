### Calcolatore serie di Fourier per funzioni periodiche.
Inserisci i tuoi dati di partenza ed esegui.

```
(*
    w = w
    T = T
    f[t_] := (Sin[w*t])^2;
*)
w = 1;
T = 2Pi/w;
f[t_] := Mod[t, 1]
```

```
a0 = Simplify[
    1/T * Integrate[f[t], {t, 0, T}]
];
an = Simplify[
    2/T * Integrate[f[t]*Cos[n*w*t], {t, 0, T}]
];
bn = Simplify[
    2/T * Integrate[f[t]*Sin[n*w*t], {t, 0, T}]
];
```

```
fn[t_, a0_, an_, bn_] := HoldForm[a0 + Sum[an*Cos[n*w*t] + bn*Sin[n*w*t], {n, 1, Infinity}]];
```

```
fn[t, a0, an, bn]
```

```
FourierSeries[f[t], t, 7]
Plot[{f[t], %}, {t, -3, 3}]
```

### Calcolatore serie di Fourier *complessa* per funzioni periodiche.
Inserisci i tuoi dati di partenza ed esegui.

```
(*
    T = 2Pi/w;
        f[t_] := (Sin[w*t])^2;
*)
w = w;
T = T;
f[t_] := 2A*t/T
```

```
cn = Simplify[
    1/T * Integrate[f[t]*Exp[-I*n*w*t], {t, 0, T}]
];
```

```
fn[t_, cn_] := HoldForm[Sum[cn*Exp[I*n*w*t], {n, -Infinity, Infinity}]];
```

```
fn[t, cn]
```
