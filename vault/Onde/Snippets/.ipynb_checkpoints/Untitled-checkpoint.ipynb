```
W = 3;
w = 1;
p = 2;
q = 0;
A = - (W^2+w^2)/W*(p + q);
L = A / (W^2 + w^2);

fx[t_] := (Cosh[w*t]*q+Sinh[w*t]*p)+ L*(W*Sinh[w*t]-w*Sin[W*t]);
fy[t_] := (Sinh[w*t]*q+Cosh[w*t]*p)+ L*(W*Cosh[w*t]-W*Cos[W*t]);
```

```
s = 0;
f = 40;
ParametricPlot[{fx[t], fy[t]}, {t, s, 1000}, WorkingPrecision->10000]
```
